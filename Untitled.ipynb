```python
import os
import sys
import json
import numpy as np

proj_base_path = '../../../'

sys.path.append(os.path.join(os.getcwd(), proj_base_path))
```

```python
with open('datasets/numpy/pokemon_clustered_time/meta/durations_dict.json') as fp:
    duration_dict = json.load(fp)
```

```python
from src.data_processing.sparse_notes_classified_time.np2mid import np2sparse
from src.generating.sparse_clustered_time_generating_seeds import seed_generators
```

```python
samples = []

for name, gen in [list(seed_generators.items())[2]]:
    x = gen(5, 128, batch_size=1)[0]
    x = np2sparse(x[:, :128], x[:, 128:], duration_dict, ppq=20)
    samples.append((name, x))
```

```python
import matplotlib.pyplot as plt

_, axs = plt.subplots(nrows=4, ncols=4, figsize=(40, 40), subplot_kw={'xticks':[], 'yticks':[]})
for ax, (name, gen) in zip(axs.flat, seed_generators.items()):
    x = gen(5, 128, batch_size=1)[0]
    x = np2sparse(x[:, :128], x[:, 128:], duration_dict, ppq=20)
    ax.imshow(x.T[::-1, :])
    ax.set_title(name)
```
